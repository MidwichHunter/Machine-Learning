Набор данных содержит список городов и количество преступлений на 100 000
человек населения по состоянию на 1970 год.

Атрибуты: 


• City – город

• Murder – количество убийств, совершенных с применением насилия.

• Rape – количество изнасилований

• Robbery – количество грабежей

• Assault – количество нападений

• Burglary – количество краж со взломом

• Larceny – количество краж

• Auto – количество угонов автомобилей.

**1. Импортируйте необходимые библиотеки и загружаем данные**

```python
# Импортируем библиотеки и загружаем данные
import numpy as np  
import pandas as pd  
import seaborn as sns  
import matplotlib.pyplot as plt  

df = pd.read_csv('crime.csv')
df.head()
```

**2. Получим информацию о датасете**

```python
df.info()
```

Избавимся от данных о городах, так как они не имеют практической пользы в анализе.

```python
cities = df['City']
df_copy = df.drop('City', axis=1)

df_copy.head(5)
```

**3. Проверяем наличие пропущенных значений и выбросов**

```python
print(df_copy.isnull().sum())
```

Пропущенные значения и категориальные признаки отсутствуют, следовательно переходим сразу к нормализации.

**4. Применяем операцию нормализации для численной устойчивости**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler
```

**5. Приведем данные к единому масштабу**

```python
X = scaler.fit_transform(df_copy)
X[:5]
```

**6. Определяем оптимальное количество кластеров методом локтя**

```python
# Определяем оптимальное количество кластеров с помощью метода локтя
from sklearn.cluster import KMeans

wcss = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(2, 11), wcss)
plt.title('Метод локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('WCSS')
plt.show()
```

На графике виден "локоть" в районе 5 кластеров. 

**7. Обучаем модель кластеризации**

```python
# заново обучим модель с тремя кластерами
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
# обучаем модель на стандартизированных данных
kmeans.fit(X)
```

**8. Предсказываем кластеры и визиализируем результаты**

```python
labels = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Кластеры уровня преступности')
plt.xlabel('Признак 0')
plt.ylabel('Признак 1')
plt.show()
```

Кластеризация успешно выявила 5 четко различимые группы. Так как один из кластеров состоит из всего 1 значения, попробуем разделить данные на 3 кластера.(другая точка перегиба в методе локтя)

```python
# заново обучим модель с тремя кластерами
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
# обучаем модель на стандартизированных данных
kmeans.fit(X)
```

```python
labels = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Кластеры уровня преступности')
plt.xlabel('Признак 0')
plt.ylabel('Признак 1')
plt.show()
```

Кластеризация успешно выявила 3 четко различимые группы, минимальное перекрытие между кластерами, что подтверждает правильность выбора количества кластеров и эффективность применения метода K-means для данного датасета.
